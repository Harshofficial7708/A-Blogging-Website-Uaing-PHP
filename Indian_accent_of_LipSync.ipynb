```
!pip install gTTS playsound pydub
import os
from gtts import gTTS
from playsound import playsound
from pydub import AudioSegment

# Create the assets directory if it doesn't exist
assets_dir = "/content/assets"
if not os.path.exists(assets_dir):
    os.makedirs(assets_dir)

# Text to convert (Indian English)
text = "Namaste Mathangi! My name is Anika, and I’m here to guide you through managing your credit card dues. Mathangi, as of today, your credit card bill shows an amount due of INR 5,053 which needs to be paid by 31st December 2024 Missing this payment could lead to two significant consequences: First, a late fee will be added to your outstanding balance. Second, your credit score will be negatively impacted, which may affect your future borrowing ability. Make your payment by clicking the link here... Pay through UPI or bank transfer. Thank you!"

# Create a gTTS object with Indian English accent
tts = gTTS(text=text, lang='en', tld='co.in')  # 'co.in' for Indian English

# Temporarily save as MP3
temp_mp3_file = os.path.join(assets_dir, "temp.mp3")
tts.save(temp_mp3_file)

# Convert MP3 to WAV using AudioSegment
audio = AudioSegment.from_mp3(temp_mp3_file)
audio_file = os.path.join(assets_dir, "indian_english_audio.wav")
audio.export(audio_file, format="wav")

# Remove the temporary MP3 file
os.remove(temp_mp3_file)

print(f"Audio saved to {audio_file}")
```

```
import os
restoredFramesPath = '/content/CodeFormer/results/images_0.7' + '/final_results/'
processedVideoOutputPath = '/content/output'

dir_list = os.listdir(restoredFramesPath)
dir_list.sort()

import cv2
import numpy as np

batch = 0
batchSize = 600
from tqdm import tqdm
for i in tqdm(range(0, len(dir_list), batchSize)):
  img_array = []
  start, end = i, i+batchSize
  print("processing ", start, end)
  for filename in  tqdm(dir_list[start:end]):
      filename = restoredFramesPath+filename;
      img = cv2.imread(filename)
      if img is None:
        continue
      height, width, layers = img.shape
      size = (width,height)
      img_array.append(img)


  out = cv2.VideoWriter(processedVideoOutputPath+'/batch_'+str(batch).zfill(4)+'.avi',cv2.VideoWriter_fourcc(*'DIVX'), 30, size)
  batch = batch + 1

  for i in range(len(img_array)):
    out.write(img_array[i])
  out.release()

# 音视频合并 Generate audio video
# Use the correct audio file path (WAV instead of MP3)
audio = ffmpeg.input(f'/content/assets/indian_english_audio.wav')  # Changed to .wav
video = ffmpeg.input(f'/content/output/batch_0000.avi')

# Specify codecs for audio and video to ensure compatibility
out = ffmpeg.output(video, audio, f'/content/output/final.mp4', vcodec='libx264', acodec='aac')  # Added codecs

# Add error handling to print ffmpeg's stderr output
try:
    out.run()
except ffmpeg.Error as e:
    print(e.stderr)
    raise e

print("output/final.mp4")
```
